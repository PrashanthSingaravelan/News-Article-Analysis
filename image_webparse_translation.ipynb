# using googletrans

```python
from googletrans import Translator
```

```python
text = 'This site is awesome'
translator = Translator()
print(translator.translate(text,dest='sw').text)
```

```python
from googletrans import Translator

text1 = "A Római Birodalom (latinul Imperium Romanum) az ókori Róma által létrehozott államalakulat volt a Földközi-tenger medencéjében"

text2 = "Vysoké Tatry sú najvyššie pohorie na Slovensku a v Poľsku a sú zároveň jediným horstvom v týchto štátoch s alpským charakterom."

translator = Translator()

dt1 = translator.detect(text1)
print(dt1)

dt2 = translator.detect(text2)
print(dt2)
```

# Using goslate

```python
import goslate
```

```python
gs = goslate.Goslate()
#translate
print(gs.translate('hello world', 'es'))
#detect
text='vanakam'
id = gs.detect(text)
print(gs.get_languages()[id])
#cant execute more times in goslate, because it results in "too many requests"
```

# Scrapping text from images

```python
from PIL import Image
from pytesseract import pytesseract

#Define path to tessaract.exe
path_to_tesseract = r'C:\Users\Sabarinathan R V\anaconda3\Library\bin\tesseract.exe'

#Define path to image
path_to_image = "test_image.png"

#Point tessaract_cmd to tessaract.exe
pytesseract.tesseract_cmd = path_to_tesseract

#Open image with PIL
pil_img = Image.open(path_to_image)

#Extract text from image
text = pytesseract.image_to_string(pil_img)

print(text)
```
